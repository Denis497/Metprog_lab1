```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sorting_times.csv")
```

```python
plt.figure()
for col in ["Heap(ms)", "Merge(ms)", "Std(ms)", "Bubble(ms)"]:
    plt.plot(data["Size"], data[col], label=col)

plt.xlabel("Размер массива")
plt.ylabel("Время сортировки (мс)")
plt.title("Сравнение алгоритмов сортировки")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

plt.figure()
for col in ["Bubble(ms)", "Merge(ms)", "Heap(ms)", "Std(ms)"]:
    plt.plot(data["Size"], np.log1p(data[col]), label=col)  # log1p(x) = log(1 + x) — работает даже для 0

plt.xlabel("Размер массива")
plt.ylabel("log(Время сортировки + 1)")
plt.title("Сравнение алгоритмов сортировки")
plt.legend()
plt.grid(True)
plt.show()
```
